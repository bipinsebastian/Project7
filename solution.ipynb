1. Graph the Number of Incidents per Location

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for incidents
data = {
    'location': ['Location A', 'Location B', 'Location C', 'Location A', 'Location B', 'Location A'],
    'count': [5, 3, 2, 4, 1, 2]
}

# Create a DataFrame
df_locations = pd.DataFrame(data)

# Group by location and count incidents
location_counts = df_locations.groupby('location')['count'].sum()

# Plotting
plt.figure(figsize=(10, 6))
location_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Incidents per Location')
plt.xlabel('Location')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

2. Graph the Number of Incidents per Day

```python
# Sample data for incidents per day
data_days = {
    'date': ['2024-10-20', '2024-10-21', '2024-10-20', '2024-10-21', '2024-10-22'],
    'count': [5, 8, 2, 3, 6]
}

# Create a DataFrame
df_days = pd.DataFrame(data_days)

# Group by date and count incidents
day_counts = df_days.groupby('date')['count'].sum()

# Plotting
plt.figure(figsize=(10, 6))
day_counts.plot(kind='line', marker='o', color='green')
plt.title('Number of Incidents per Day')
plt.xlabel('Date')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

3. Graph the Number of Incidents per Hour in the Last 24

```python
# Sample data for incidents per hour
data_hours = {
    'hour': ['00', '01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12', '13', '14', '15'],
    'count': [1, 2, 0, 3, 1, 2, 5, 1, 0, 3, 2, 4, 0, 2, 1, 1]
}

# Create a DataFrame
df_hours = pd.DataFrame(data_hours)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df_hours['hour'], df_hours['count'], color='orange')
plt.title('Number of Incidents per Hour in the Last 24 Hours')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

4. Graph the Number of Incidents per Event Type

```python
# Sample data for incidents per event type
data_event_types = {
    'event_type': ['Type A', 'Type B', 'Type A', 'Type C', 'Type B', 'Type A', 'Type C'],
    'count': [3, 2, 5, 1, 4, 2, 1]
}

# Create a DataFrame
df_event_types = pd.DataFrame(data_event_types)

# Group by event type and count incidents
event_type_counts = df_event_types.groupby('event_type')['count'].sum()

# Plotting
plt.figure(figsize=(10, 6))
event_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Number of Incidents per Event Type')
plt.ylabel('')
plt.grid(axis='y')
plt.show()
```

5. Map the Locations Where Incidents Have Occurred

```python
import folium

# Sample data for incident locations with counts
incident_locations = [
    {'name': 'Location A', 'latitude': 59.3293, 'longitude': 18.0686, 'count': 10},  # Stockholm
    {'name': 'Location B', 'latitude': 55.6050, 'longitude': 13.0038, 'count': 5},   # Malmö
    {'name': 'Location C', 'latitude': 57.7089, 'longitude': 11.9746, 'count': 3}    # Gothenburg
]

# Initialize the map
incident_map = folium.Map(location=[61.0, 14.0], zoom_start=5)

# Add markers to the map
for location in incident_locations:
    folium.CircleMarker(
        location=(location['latitude'], location['longitude']),
        radius=location['count'],
        color='blue',
        fill=True,
        fill_opacity=0.6,
        popup=f"{location['name']}: {location['count']} incidents"
    ).add_to(incident_map)

# Save the map to an HTML file
incident_map.save('incident_map.html')
```
